```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
sns.set() # Set default plotter to be seaborn
```

```python
data = sns.load_dataset('planets') 
data=pd.DataFrame(data)
data = data[data["method"] == "Radial Velocity"]
data
```

### This is the relationship between the mass of planet and the Radial velocity of planets, as we show when the mass incresed the radial velocity will decrease. So the plantes that have less masses have a less radial velocity.

```python
sns.histplot(data=data, x="mass", hue="method", multiple="stack")
```

## this flow chart shows the relation ship between the orbited period and distance of planets, as we show when the distance increase the orbital period decrease.                                                          

```python
sns.histplot(data=data.head(30), x="orbital_period", hue="distance", multiple="stack")
```

## This flow chart indicate that there is aleinar relationship between the orbital periode, when mass incresed the orbital periode will  increase 

```python
sns.regplot(x='mass', y='orbital_period', data=data, color='blue')
```
